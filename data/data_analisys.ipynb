```python
import seaborn as sns
import pandas as pd
```

```python
dataset = pd.read_csv("./dataset.csv")
dataset
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.categorical.countplot(x=dataset["sentiment"])
```

```python
just_labels = dataset.loc[dataset.sentiment != "no"]
sns.categorical.countplot(x=just_labels["sentiment"])
```

```python
sns.set(font_scale=0)
sns.categorical.countplot(x=just_labels["aspect"])
```

```python
sns.set(font_scale=1)
no_labels = dataset.loc[dataset.sentiment == "no"]
new_dataset = pd.concat([no_labels.sample(5000), just_labels])
sns.categorical.countplot(x=new_dataset["sentiment"])
```

```python
sns.set(font_scale=0)
sns.categorical.countplot(x=new_dataset["aspect"])
```
